```python
#configuring piplines for the classifiers
## Naive Base classifier 
def MultinomialNB_train(X,y):
    pipelineNB = Pipeline([
        ('vect', TfidfVectorizer()),
        ('norm', preprocessing.Normalizer(norm='l2')),
        ('clf_nb', MultinomialNB()),
    ])
    return pipelineNB.fit(X,y)
    
## Perceptron classifier
def Perceptron_train(X,y):
    pipelinePR = Pipeline([
        ('vect', TfidfVectorizer()),
        ('norm', preprocessing.Normalizer(norm='l2')),
        ('clf',Perceptron(tol=1e-3, random_state=42, alpha=0.00001, max_iter=10))
    ])
    return pipelinePR.fit(X,y)

## SVC classifier 
def linearSVC_train(x_train,y_train):
    pipelineSVC = Pipeline([
        ('vect', TfidfVectorizer()),
        ('norm', preprocessing.Normalizer(norm='l2')),
        ('clf',LinearSVC(class_weight ='balanced'))
    ])
    return pipelineSVC.fit(x_train,y_train)

def KNN_train(x_train,y_train,k=3):
    pipelineKNN = Pipeline([
        ('vect', TfidfVectorizer()),
        ('norm', preprocessing.Normalizer(norm='l2')),
        ('clf',KNeighborsClassifier(n_neighbors=k))
    ])
    return pipelineKNN.fit(x_train,y_train)

```

```python
#Processing the Data 
## removing digits from the story ##
def delete_digits(story):
    translator  = str.maketrans("", "", string.digits)
    return story.translate(translator)

## removing english from the story ##
def delete_english(story):
    return re.sub('[a-zA-Z]+', '', story)

## removing english from the story ##
def delete_punctuation(story):
    translator = str.maketrans("", "", string.punctuation)
    return story.translate(translator)


## seraching heb words ##
def Tokenizing(dataframe):
    story  = ' '.join(dataframe)
    tokens = ht.tokenize(story)
    words_list = []
    for tuple in tokens:
        words_list.append(tuple[1])
    return words_list
    

## Removing words any words that we have not deleted yet ## 
def delete_words(stories,tokens):
    for i in range(len(stories)):
        processed_story = ''
        for word in stories[i].split():
            if word in tokens:
                processed_story+= ' ' + word
        stories[i] = processed_story
    stories               
    return stories
```

```python
clean_df = df_train.copy()
clean_df['story'] = clean_df['story'].apply(delete_punctuation)# deleteing Punctuation
clean_df['story'] = clean_df['story'].apply(delete_digits)# Deleting digits
clean_df['story'] = clean_df['story'].apply(delete_english)# Deleting english letters
tokens = list(Tokenizing(clean_df['story']))
tokens = list(dict.fromkeys(tokens))#deleting duplicate words from the tokens list
clean_df['story'] = delete_words(clean_df['story'],tokens)
```

# Testing Tokenization, Vectorization and Classifiers

```python
#Copying the dataframes for testing
X = clean_df['story'].copy()
y = clean_df['gender'].copy()
X_test =  df_test['story'].copy()
```

```python
#Testing Converting into Count Vector
text_samples = clean_df['story'].values[:10]
vec = CountVectorizer()
X_train = vec.fit_transform(text_samples)
count_vec_df = pd.DataFrame(X_train.toarray(), columns=vec.get_feature_names_out())
count_vec_df.shape
count_vec_df

```

```python
#Testing Converting into Tfidf Vector
vec = TfidfVectorizer()
X_train_tfidf = vec.fit_transform(text_samples)
tfidf_vec_df = pd.DataFrame(X_train_tfidf.toarray(), columns=vec.get_feature_names_out())
tfidf_vec_df.shape
tfidf_vec_df
```

```python
## NaiveBase prediction with different Alphas
clf = MultinomialNB_train(X,y)
f1_macro = cross_val_score(clf, X, y, cv=10,  scoring='f1_macro').mean()
print("The f1 macro score is for NaiveBase is :\n{}".format(f1_macro))
```

```python
## Perceptron prediction
clf = Perceptron_train(X,y)
f1_macro = cross_val_score(clf, X, y, cv=10,  scoring='f1_macro').mean()
print("The f1 macro score is for Perceptron is :\n{}".format(f1_macro))
```

```python
### SVM prediction
clf = linearSVC_train(X,y)
f1_macro = cross_val_score(clf, X, y, cv=10,  scoring='f1_macro').mean()
print("The f1 macro score is for SVM is :\n{}".format(f1_macro))
```

```python
### KNN prediction
clf = KNN_train(X,y)
for k in range(3,10):
    clf = KNN_train(X,y,k)
    f1_macro = cross_val_score(clf, X, y, cv=10,  scoring='f1_macro').mean()
    print("for k = {}".format(k))
    print("The f1 macro score is for KNN is :\n{}".format(f1_macro))
```

# The best f1_macro score is Perceptron

```python
#using Perceptron on the test dataset
clf = Perceptron_train(X,y)
prediction = clf.predict(X_test)
df_predicted = pd.DataFrame({"test_example_id" : df_test["test_example_id"] ,"predicted_category" : prediction })
df_predicted
```
